# Matplotlib

```python
from matplotlib import pylab
pylab.__version__
```

### Numpy 

```python
# Using numpy to generate random sample data
import numpy as np

x = np.linspace(0,10,25)
y = x * x + 2
print(x)
print(y)

# Reshaping array
print(">>>")
print(np.array([x, y]).reshape(25,2))
print(">>>")
print(np.array([x, y]).reshape(2,25))
print(">>>")
print(np.array([x, y]).reshape(25,2).reshape(2,25))
```

### Simple Graph

```python
# Plotting on graph
pylab.plot(x, y, 'r') # r stand for color of line
```

### Subgraph

```python
# Two side by side graph
pylab.subplot(1,2,1) # (rows, columns, indexes)
pylab.plot(x, y, 'r--')
pylab.subplot(1,2,2)
pylab.plot(y,x, 'g*-')
```

```python
# Both graph in same plot
pylab.subplot(1,1,1) # (rows, columns, indexes)
pylab.plot(x, y, 'r--')
pylab.subplot(1,1,1)
pylab.plot(y,x, 'g*-')
```

### Operation Description

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axes = fig.add_axes([0.5, 0.1, 0.8, 0.8]) # Control the left, rigth, width, height of the canvas from 0 to 1

axes.plot(x,y, 'r--')
```

### Subgraph

```python
fig, axes = plt.subplots(nrows=2, ncols=1) # sub mao is 1 rows and 2 columns

for ax in axes :
    ax.plot(x, y, 'r--')
```

### A graph within graph

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

axes1.plot(x, y, 'g*-')
axes2.plot(y, x, 'r--')
```

### Add Graphic to the graph

```python
fig = plt.figure(figsize=(16,9), dpi=300)

fig.add_subplot()
plt.plot(x, y, 'r--')
```

### Labeling the Graph

```python
fig, axes = plt.subplots()
axes.set_title('Graph Title')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.legend(['y = x**2', 'y = x**3'], loc=0)

axes.plot(x, x**2)
axes.plot(x, x**3)
```

### Line Color, transparency, size, style, and more

```python
# Color and transparency
fig, axes = plt.subplots(dpi=250)
axes.plot(x, x**1.2, color='red', alpha=.6)
axes.plot(x, x+2, color='blue', alpha=.6)
axes.plot(x, x+3, color='green', alpha=.6)
```

```python
# line width
fig, axes = plt.subplots(dpi=250)
axes.plot(x, x+1, color='red', linewidth=.25)
axes.plot(x, x+2, color='red', linewidth=1)
axes.plot(x, x+3, color='red', lw=1.5)
axes.plot(x, x+4, color='red', lw=2)
```

```python
# Line Style
fig, axes = plt.subplots(dpi=250)
axes.plot(x, x+1, color='red', lw=1.24, linestyle='-')
axes.plot(x, x+2, color='red', lw=1.2, linestyle='--')
axes.plot(x, x+3, color='red', lw=1.2, ls='-.')
axes.plot(x, x+4, color='red', lw=1.2, ls=':')

line, = axes.plot(x, x+5, color='black', lw=1.2)
line.set_dashes([5,10,15,10])

line, = axes.plot(x, x+6, color='black', lw=1.2)
line.set_dashes([5,10,15,3])

line, = axes.plot(x, x+7, color='black', lw=1.2)
line.set_dashes([5,10,4,10])
```

```python
# adding symbols for data in line
fig, axes = plt.subplots(dpi=250)
axes.plot(x, x+1, color='red', lw=1, marker='o', markersize=3)
axes.plot(x, x+2, color='red', lw=1, marker='+')
axes.plot(x, x+3, color='red', lw=1, marker='*')
axes.plot(x, x+4, color='red', lw=1, marker='s', markerfacecolor='blue')
axes.plot(x, x+5, color='red', lw=1, marker='v')
axes.plot(x, x+6, color='red', lw=1, marker='p')
axes.plot(x, x+7, color='red', lw=1, marker='h')
```

### Setting axis range

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].plot(x, x**2, x, x**3, lw=1.5)
axes[0].grid(True)

axes[1].plot(x, x**2, x, x**3, lw=1.5)
axes[1].grid(True)
axes[1].set_ylim([0, 2])
axes[1].set_xlim([0,1.25])
```

### Other 2D Graphics

```python
n = np.array([0, 1, 2, 3, 4, 5])

fig, axes = plt.subplots(1, 4, figsize=(16,5))

# Scatter Plot
axes[0].set_title('Scatter Plot')
axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))

# Step Plot
axes[1].set_title('Step Plot')
axes[1].step(n, n**2, lw=2)

# bar Graph
axes[2].set_title('Bar Graph')
axes[2].bar(n, n**2, align='center', alpha=0.8)

# Fill Betweens
axes[3].set_title('Fill Between Graph')
axes[3].fill_between(x, x**2, x**3, color='green', alpha=0.6)
```

```python
# Radar Chart

fig = plt.figure(figsize=(6,6))
axes = fig.add_axes([0.0, 0.0, .6, .6], polar=True)

t = np.linspace(0, 2* np.pi, 100)
axes.plot(t,t, lw=3)
```

```python
# Histogram

n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))
axes[0].set_title('Histogram Graph')
axes[0].hist(n)

axes[1].set_title('Cumulative Histogram')
axes[1].hist(n, cumulative=True, bins=8)
```

### Contour Image

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2
```

```python
fig, axes = plt.subplots()
cs = axes.contour(X, Y, Z)
axes.clabel(cs, inline=1, fontsize=10)
axes.set_title('contour map')
```

### 3D Surface

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize=(14,6))

axes = fig.add_subplot(1, 2, 1, projection='3d')
axes.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)
```

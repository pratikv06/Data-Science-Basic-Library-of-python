<div class="alert alert-block alert-info">
    <h1> 1. How to read few data from columns and row from CSV? </h1>
</div>

```python
movies = pd.read_csv('http://bit.ly/imdbratings', usecols=[1,3], nrows= 10)
movies
```

<div class="alert alert-block alert-info">
<h1> 2. Using for loop on dataframe </h1>
</div>

```python
for movie in movies:
    print(movie)
```

```python
for movie in movies.title:
    print(movie)
```

```python
for movie in movies.genre:
    print(movie)
```

```python
for index, row in movies.iterrows():
    print(index, row.title, row.genre)
```

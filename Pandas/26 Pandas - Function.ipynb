# How do I apply a function to a pandas seroes or dataframe

```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
# Adding a new column in the dataframe
train['Sex_num'] = train.Sex.map({'female':0, 'male':1})
train.loc[0:5, ['Sex', 'Sex_num']]
```

```python
# Calculating name length in new column
train['Name_length'] = train.Name.apply(len)
train.loc[0:4, ['Name', 'Name_length']]
```

```python
# using numpy function
import numpy as np
train['Fare_ceil'] = train.Fare.apply(np.ceil)
train.loc[0:4, ['Fare', 'Fare_ceil']]
```

```python
# fetching firat name from Name column
train.Name.str.split(',').head()
```

```python
# Method 1: Uding user defined function
def get_element(mylist, position):
    return mylist[position]
```

```python
train.Name.str.split(',').apply(get_element, position=0).head()
```

```python
# Method 2: Using Lan=mbda Function
train.Name.str.split(',').apply(lambda x :x[0]).head()
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
drinks.loc[:, 'beer_servings': 'wine_servings']
```

```python
# Maximum of each column
drinks.loc[:, 'beer_servings': 'wine_servings'].apply(max, axis=0)
```

```python
# Maximum form each row
drinks.loc[:, 'beer_servings': 'wine_servings'].apply(max, axis=1)
```

```python
# Maximum column name
drinks.loc[:, 'beer_servings': 'wine_servings'].apply(np.argmax, axis=1)
```

```python
# Changing the data type of each column to float
drinks.loc[:, 'beer_servings': 'wine_servings'].applymap(float)
```

```python
# making changes in original dataframe
drinks.loc[:, 'beer_servings': 'wine_servings'] = drinks.loc[:, 'beer_servings': 'wine_servings'].applymap(float)
drinks.head()
```

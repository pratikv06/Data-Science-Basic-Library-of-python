# How do I use the "axis" parameter in pandas 

```python
import pandas as pd
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
drinks[drinks['country'] == 'India']
```

```python
drinks.drop(['continent'], axis=1).head()
```

```python
drinks.drop(2, axis=0).head()
```

```python
# default axis = 0
# add all the rows, and take columns as header
drinks.mean()
```

```python
# add all the columns, and take row as header
drinks.mean(axis=1)
```

```python
print(drinks.shape)
print(drinks.mean().shape)
print(drinks.mean(axis=1).shape)
```

# How do I sort a pandas DataFrame or a series? 

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
# Only One Column at a time, and display only that column
print(movies.title.sort_values())
print(type(movies.title.sort_values()))
print('*-*'*25)
print(movies['title'].sort_values(ascending=False))
print(type(movies['title'].sort_values(ascending=False)))
```

```python
# Sort and display all column
movies.sort_values('duration', ascending=False, inplace=True)
movies.head()
```

```python
# Sort multiple columnsand display all column
movies.sort_values(['duration', 'star_rating'])
```

```python
# Old Method of sorting
movies.title.order()
```

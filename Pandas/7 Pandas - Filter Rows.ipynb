# How do I filter rows of a pandas DataFrame by Columns values?

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
type(movies.title)
```

```python
# METHOD 1
boolean = []
for length in movies.duration:
    if length >= 200:
        boolean.append(True)
    else:
        boolean.append(False)
```

```python
boolean[0:5]
```

```python
# Converting list to pandas series
# `S` of series is capital
is_long = pd.Series(boolean)
is_long.head()
```

```python
movies[is_long]
```

```python
# METHOD 2
is_long1 = movies.duration >= 200
movies[is_long1]
```

```python
# METHOD 3
movies[movies['duration'] >= 200]
```

```python
# Extra
movies[movies['duration'] >= 200]['genre']
```

```python
movies.loc[movies['duration'] >= 200, 'genre']
```

```python
movies.loc[movies['duration'] >= 200, ('star_rating', 'genre')]
```

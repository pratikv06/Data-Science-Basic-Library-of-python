# How do I select multiple columns and row from a Pandas DataFrame? 

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

# loc 

```python
# one row, all column
ufo.loc[0, :]
```

```python
# multiple row and all column
# ufo.loc[[0, 1, 2], :]
ufo.loc[0:2,:]
```

```python
# all rows, and multiple columns
# ufo[['City', 'State']] - Same as below statemenet
ufo.loc[:, ['City', 'State']]
```

```python
ufo[ufo.City=='Oakland']
```

```python
# using loc to et above result
ufo.loc[ufo.City=='Oakland', :]
```

# iloc

in loc - 1:4 [means both are inclusive] <br>
in iloc - 1:4 [means before 4]

```python
ufo.iloc[:, [0,3]]
```

```python
# ufo[0:2] - Same as below statement
ufo.iloc[0:2,:]
```

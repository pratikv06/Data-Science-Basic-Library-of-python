# How do I use pandas with scikit-learn to create Kaggle submission? 

```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
feature_cols = ['Pclass', 'Parch']
X = train.loc[:, feature_cols]
X.shape
```

```python
Y = train.Survived
Y.shape
```

## scikit-learn Model

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X,Y)
```

```python
test = pd.read_csv('http://bit.ly/kaggletest')
test.head()
# survive columns is missing, we are going to predict it
```

```python
X_new = test.loc[:, feature_cols]
X_new.shape
```

```python
new_pred_class = logreg.predict(X_new)
new_pred_class
```

```python
pd.DataFrame({'PassengerID': test.PassengerId, 'Survived': new_pred_class}).set_index('PassengerID').to_csv('19_sub.csv')
```

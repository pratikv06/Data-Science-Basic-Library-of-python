# How do I remove columns from a pandas dataframe? 

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
# Drop method 
# axis, 1 = Columns, 0 = Rows
ufo.drop('Colors Reported', axis=1, inplace=True)
ufo.head()
```

```python
# Multiple Drop
ufo.drop(['City', 'State'], axis=1, inplace=True)
ufo.head()
```

```python
# Droping data from rows
# Index is the header for row
ufo.drop([0, 1], axis=0, inplace=True)
ufo.head()
```

```python
# Question :
# Resetting the index of row to 0 after deleting row data
ufo.reset_index(drop=True, inplace=True)
ufo.head()
```

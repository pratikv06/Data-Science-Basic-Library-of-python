# How do I create dummy variables in pandas?

```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
train['Sex Male'] = train.Sex.map({'female':0, 'male':1})
train.head()
```

```python
# create 0's/1's dataframe of all possible value in column
pd.get_dummies(train.Sex).head(8)
```

```python
# Now removing female column and adding prefix to column name
pd.get_dummies(train.Sex, prefix='Sex').iloc[:, 1:].head(8)
```

```python
# Doing the same with 3 possible set in column and adding this value in original dataframe
embarked_dummies = pd.get_dummies(train.Embarked, prefix='Embarked').iloc[:,1:]
train = pd.concat([train, embarked_dummies], axis=1)
train.head()
```

```python
# This thing can be done in one line
train1 = pd.read_csv('http://bit.ly/kaggletrain')
pd.get_dummies(train1, columns=['Sex', 'Embarked'], drop_first=True)
```
